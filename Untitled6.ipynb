```
from PIL import Image
import numpy as np
from collections import deque
```

```
im = Image.open('sample.tif')
arr = np.array(im)
```

```
n, m = arr.shape
x, y = [int(x) for x in input('Enter coordinates : ').split()]
vis = [[0 for i in range(m)] for j in range(n)] 
s  = deque()
```

```
def check_boundary(x, y):
  if (x >= 0 and x < n and y >= 0 and y < m):
    return True
  return False
```

```
def add_operation(x_1, y_1, x_2, y_2):
  print(str(x_1) +' '+str(y_1) +' --> '+str(x_2) +' '+str(y_2))
  vis[x_2][y_2] = 1
  arr[x_2][y_2] = arr[x_2][y_2] + arr[x_1][y_1]
  s.append([x_2, y_2])
```

```
def algo(x, y):
  # check upper left diag
	if (check_boundary(x - 1, y - 1) and vis[x - 1][y - 1] != 1):
		add_operation(x, y, x - 1, y - 1)
  
		if (check_boundary(x - 1, y - 2) and vis[x - 1][y - 2] != 1):
			add_operation(x - 1, y - 1, x - 1, y - 2)

		if (check_boundary(x - 2, y - 1) and vis[x - 2][y - 1] != 1):
			add_operation(x - 1, y - 1, x - 2, y - 1)

		if (check_boundary(x - 2, y - 2) and vis[x - 2][y - 2] != 1):
			add_operation(x - 1, y - 1, x - 2, y - 2)

	# check lower left diag
	if (check_boundary(x + 1, y - 1) and vis[x + 1][y - 1] != 1):
   
		add_operation(x, y, x + 1, y - 1)
  
		if (check_boundary(x + 1, y - 2) and vis[x + 1][y - 2]!= 1):
			add_operation(x + 1, y - 1, x + 1, y - 2)

		if (check_boundary(x + 2, y - 1) and vis[x + 2][y - 1]!= 1):
			add_operation(x + 1, y - 1, x + 2, y - 1)

		if (check_boundary(x + 2, y - 2) and vis[x + 2][y - 2]!= 1):
			add_operation(x + 1, y - 1, x + 2, y - 2)

	# check upper right diag
	if (check_boundary(x - 1, y + 1) and vis[x - 1][y + 1]!= 1):
		add_operation(x, y, x - 1, y + 1)

		if (check_boundary(x - 1, y + 2) and vis[x - 1][y + 2]!= 1):
			add_operation(x - 1, y + 1, x - 1, y + 2)

		if (check_boundary(x - 2, y + 1) and vis[x - 2][y + 1]!= 1):
			add_operation(x - 1, y + 1, x - 2, y + 1)

		if (check_boundary(x - 2, y + 2) and vis[x - 2][y + 2]!= 1):
			add_operation(x - 1, y + 1, x - 2, y + 2)

	# check lower right diag
	if (check_boundary(x + 1, y + 1) and vis[x + 1][y + 1]!= 1):

		add_operation(x, y, x + 1, y + 1)

		if (check_boundary(x + 2, y + 1) and vis[x + 2][y + 1]!= 1):
			add_operation(x + 1, y + 1, x + 2, y + 1)

		if (check_boundary(x + 1, y + 2) and vis[x + 1][y + 2]!= 1):
			add_operation(x + 1, y + 1, x + 1, y + 2)

		if (check_boundary(x + 2, y + 2) and vis[x + 2][y + 2]!= 1):
			add_operation(x + 1, y + 1, x + 2, y + 2)

	# check top
	if (check_boundary(x - 1, y) and vis[x - 1][y]!= 1):
		add_operation(x, y, x - 1, y)

	# check bottom
	if (check_boundary(x + 1, y) and vis[x + 1][y]!= 1):
		add_operation(x, y, x + 1, y)

	# check left
	if (check_boundary(x, y - 1) and vis[x][y - 1]!= 1):
		add_operation(x, y, x, y - 1)

	# check right
	if (check_boundary(x, y + 1) and vis[x][y + 1] != 1):
		add_operation(x, y, x, y + 1)
```

```
vis[x][y] = 1
s.append([x, y])
while(len(s) > 0):
  cur_x = s[-1][0]
  cur_y = s[-1][1]
  s.pop()
  algo(cur_x, cur_y)
print(arr)
```
